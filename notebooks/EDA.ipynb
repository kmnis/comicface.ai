```python
import os
import numpy as np
from glob import glob

import matplotlib.pyplot as plt
from matplotlib.image import imread

from src.data_loader import preprocess_train_image
from tensorflow.keras.utils import array_to_img
```

```python
image_files = glob("../data/**/*.jpg")
image_files.sort()
len(image_files)
```

```python
num_samples = 5  # Number of random samples to display
hist_bins = 256  # Number of bins for histogram

fig, axes = plt.subplots(3, num_samples, figsize=(15, 8))
for i in range(num_samples):
    ind = np.random.randint(10000, 20000)
    processed_images = preprocess_train_image(image_files[ind], image_files[ind-10000])
    
    image = plt.imread(image_files[ind])
    axes[0, i].imshow(image)
    axes[0, i].set_title(f"Sample {i + 1}")
    axes[0, i].axis('off')
    
    axes[1, i].imshow(array_to_img(processed_images[0]))
    axes[1, i].axis('off')
    
    axes[2, i].imshow(array_to_img(processed_images[1]))
    axes[2, i].axis('off')

axes[0, 0].set_ylabel("Original Image")
axes[1, 0].set_ylabel("Processed Image")
axes[2, 0].set_ylabel("Processed Comic Image")

plt.tight_layout()
plt.show()
```
